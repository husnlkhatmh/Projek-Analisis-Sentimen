```python
from google_play_scraper import reviews, Sort
import pandas as pd
import csv
```

```python
ulasan, _ = reviews(
    'id.co.bri.brimo',          # ID aplikasi BRImo di Play Store
    lang='id',                  # Ambil ulasan dalam Bahasa Indonesia
    country='id',               # Target dari pengguna di Indonesia
    sort=Sort.MOST_RELEVANT,    # Urutkan berdasarkan ulasan yang paling relevan
    count=10000                 # Batas min jumlah ulasan yang ingin diambil
)

data_ulasan = pd.DataFrame(ulasan)
data_ulasan.to_csv('brimo_dataset.csv', index=False)
```
